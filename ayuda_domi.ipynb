```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_ventas = "datos/Ventas por Cliente/ventas_con_precio_lista_y_descuentos_2025_2026.csv"
ventas = pd.read_csv(path_ventas)
ventas.head()
```

```python
ventas.columns
```

```python
path_info_zonales = "datos/Clientes - Zonales.xlsx"
zonales = pd.read_excel(path_info_zonales)
zonales.head()
```

```python
zonales.columns
```

```python
ventas = ventas.merge(zonales[["Id cliente", "REGIONES"]].rename(columns={"Id cliente": "Cod Cliente", "REGIONES": "zonal"}), on="Cod Cliente", how="left")
```

```python
ventas.columns
```

```python
# Filtrar solo los canales relevantes
canales_relevantes = [
    "COBERTURA",
    "VOLUMEN COBERTURA",
    # "MAYORISTAS CADENAS",
    # "MAYORISTA B VOLUMEN",
    # "OTROS MAYORISTAS",
    # "HORECA VOLUMEN",
    # "OTROS HORECA",
]
ventas = ventas[
    (ventas["Cod Canal Comercial"] == "CB") &
    (ventas["Año"] == 2025)
]
ventas = ventas[ventas["Nombre Consolidado"].isin(canales_relevantes)]
print("Ventas después de filtrar canales irrelevantes - Filas:", ventas.shape[0])
ventas.head()
```

```python
tabla_domi = ventas.groupby(["Año", "zonal", "Cod Cliente", "Nombre Familia", "Nombre Marca", "Cod SKU", "Nombre SKU"]).agg({
    "Monto Real": "sum"
}).reset_index()
tabla_domi.head()
```

```python
tabla_domi.to_excel("tabla_domi.xlsx")
```
