```python
import pandas as pd
```

```python
path_segmentos_sku = "datos/Propuesta de productos a eliminar.xlsx"
segmentos_sku = pd.read_excel(path_segmentos_sku, sheet_name="BBDD")
segmentos_sku.head()
```

```python
segmentos_sku.columns
```

```python
path_infaltables = "datos/Base Infaltables.xlsx"
infaltables = pd.read_excel(path_infaltables)
infaltables.head()
```

```python
infaltables.columns
```

```python
infaltables = infaltables[
    (infaltables["Canal"]=="CB") &
    (infaltables["Tipo SKU"]=="Infaltable")
]
```

```python
skus_infaltables = infaltables["CodSKU"].unique()
print(len(skus_infaltables))
print(skus_infaltables)
```

```python
segmentos_sku["infaltable"] = segmentos_sku["SKU"].isin(skus_infaltables)
```

```python
# Como se distribuyen los infaltables en los segmentos
segmentos_sku[
    (segmentos_sku["infaltable"])
]["Segmento Producto"].value_counts()
```

```python
# Como se distribuyen los heroes en los infaltables
segmentos_sku[
    (segmentos_sku["Segmento Producto"]=="Héroe")
]["infaltable"].value_counts()
```

```python
tabla_segmentacion_infaltables = (
    pd.crosstab(segmentos_sku["Segmento Producto"], segmentos_sku["infaltable"])
    .rename(columns={True: "Infaltable", False: "No infaltable"})
    .reindex(columns=["Infaltable", "No infaltable"], fill_value=0)
)

tabla_segmentacion_infaltables
```

Se han definido Productos Homologables, de manera de no afectar la ejecución en aquellos puntos de 
venta que no trabajan el formato exigible de los productos infaltables. 

```python
ingresos_heroes = segmentos_sku[
    (segmentos_sku["Segmento Producto"]=="Héroe")
]["Ingresos ('000 CLP)"].sum()
ingresos_heroes_infaltables = segmentos_sku[
    (segmentos_sku["Segmento Producto"]=="Héroe") &
    (segmentos_sku["infaltable"])
]["Ingresos ('000 CLP)"].sum()

print(f"Ingresos Heroes infaltables: {ingresos_heroes_infaltables}")
print(f"Ingresos Heroes: {ingresos_heroes}")
print(f"Esto significa un {ingresos_heroes_infaltables/ingresos_heroes * 100}%")
```

```python
ebitda_heroes = segmentos_sku[
    (segmentos_sku["Segmento Producto"]=="Héroe")
]["EBITDA ('000 CLP)"].sum()
ebitda_heroes_infaltables = segmentos_sku[
    (segmentos_sku["Segmento Producto"]=="Héroe") &
    (segmentos_sku["infaltable"])
]["EBITDA ('000 CLP)"].sum()

print(f"Ingresos Heroes infaltables: {ebitda_heroes_infaltables}")
print(f"Ingresos Heroes: {ebitda_heroes}")
print(f"Esto significa un {ebitda_heroes_infaltables/ebitda_heroes * 100}%")
```
