```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path_ventas_2026 = "datos/Ventas por Cliente/Ventas 2026/Base Ventas x Facturas 2026.xlsx"
ventas_2026_2 = pd.read_excel(path_ventas_2026, header=0, sheet_name=1)
ventas_2026_3 = pd.read_excel(path_ventas_2026, header=0, sheet_name=2)

ventas_2026_pendientes = pd.concat([ventas_2026_2, ventas_2026_3], ignore_index=True)
```

```python
ventas_2026_pendientes = ventas_2026_pendientes.rename(columns={
    "Unnamed: 0": "Índice",
    "Monto Real": "Monto Real",
    "Monto Nc": "Monto Nc",
    "Kilos Nc": "Kilos Nc"
})
ventas_2026_pendientes = ventas_2026_pendientes.drop(columns=["Índice"])
ventas_2026_pendientes.columns
```

```python
ventas = pd.read_csv("datos/Ventas por Cliente/ventas_2025_2026.csv")
ventas.columns
```

```python
ventas_new = pd.concat([ventas, ventas_2026_pendientes], ignore_index=True)
```

```python
ventas_new.to_csv("datos/Ventas por Cliente/ventas_2025_2026.csv", index=False)
```
