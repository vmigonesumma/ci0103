```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Cargar datos
# path_ventas = "datos/Ventas por Cliente/ventas_con_precio_lista_y_descuentos_2025_2026.csv"
path_ventas = "datos/Ventas por Cliente/ventas_2025_2026.csv"
ventas = pd.read_csv(path_ventas)
print("Ventas - Filas:", ventas.shape[0])
ventas.head()
```

```python
ventas.columns
```

```python
# path_info_zonales = "datos/Clientes - Zonales.xlsx"
path_info_zonales = "Cliente Zonal anual.xlsx"
zonales = pd.read_excel(path_info_zonales)
zonales.head()
```

```python
zonales.groupby("Id cliente").agg({
    "REGIONES": "nunique"
}).max()
```

```python
zonales.columns
```

```python
mapeo_cliente_zonal = dict(zip(zonales['Id cliente'], zonales['REGIONES']))
```

```python
## Filtrar solo los canales relevantes
canales_relevantes = [
    "COBERTURA",
    "VOLUMEN COBERTURA",
    # "MAYORISTAS CADENAS",
    # "MAYORISTA B VOLUMEN",
    # "OTROS MAYORISTAS",
    # "HORECA VOLUMEN",
    # "OTROS HORECA",
]
ventas = ventas[ventas["Nombre Consolidado"].isin(canales_relevantes)]
print("Ventas después de filtrar canales irrelevantes - Filas:", ventas.shape[0])

ventas = ventas[
    ((ventas["Año"] == 2025) & (ventas["Mes"] >= 10)) |
    (ventas["Año"] == 2026)
]
print("Ventas después de filtrar fechas - Filas:", ventas.shape[0])
ventas.head()
```

```python
# año mes en formato oct-25
ventas["Año2"] = ventas["Año"].map({2025:25, 2026:26})
ventas["Mes2"] = ventas["Mes"].map({
    10: "oct",
    11: "nov",
    12: "dic",
    1:  "ene",
    2:  "feb",
    3:  "mar"
})
ventas["año_mes"] = ventas["Mes2"].astype(str) + "-" + ventas["Año2"].astype(str)
```

```python
mapeo_cliente_consolidado = dict(zip(ventas['Cod Cliente'], ventas['Nombre Consolidado']))
```

```python
base_ventas_para_costeo = ventas.groupby([
    "año_mes",
    "Cod Cliente",
    "Cod SKU"
]).agg({
    "Kilo Real": "sum",
    "Monto Real": "sum",
}).reset_index(
).rename(columns={
    "Cod Cliente": "cod_cliente",
    "Cod SKU": "cod_sku",
    "Kilo Real": "kg_totales",
    "Monto Real": "facturacion",
})

base_ventas_para_costeo["zonal"] = base_ventas_para_costeo["cod_cliente"].map(mapeo_cliente_zonal)
base_ventas_para_costeo["consolidado"] = base_ventas_para_costeo["cod_cliente"].map(mapeo_cliente_consolidado)

base_ventas_para_costeo.head()
```

```python
base_ventas_para_costeo = base_ventas_para_costeo[[
    "año_mes",
    "cod_cliente",
    "zonal",
    "consolidado",
    "cod_sku",
    "kg_totales",
    "facturacion"
]]

base_ventas_para_costeo.head()
```
