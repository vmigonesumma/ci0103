```python
# --- 1. Preparar tipos de dato ---
# Ajusta el formato de fecha según tu CSV (ej: '%d.%m.%Y', '%Y-%m-%d', etc.)
# En ventas está 20260106
date_fmt_ventas = '%Y%m%d'
# En precios lista está 2023-04-19
date_fmt_precios = '%Y-%m-%d'

ventas['Fecha Factura'] = pd.to_datetime(ventas['Fecha Factura'], format=date_fmt_ventas)
precios_lista['Válido de'] = pd.to_datetime(precios_lista['Válido de'], format=date_fmt_precios)
precios_lista['Fin de validez'] = pd.to_datetime(precios_lista['Fin de validez'], format=date_fmt_precios)
```

```python
# Asegurar que las keys de join sean del mismo tipo
ventas = ventas.copy()
ventas['venta_id'] = ventas.index

ventas['Cod SKU'] = ventas['Cod SKU'].astype(str).str.strip()
ventas['Cod Canal Comercial'] = ventas['Cod Canal Comercial'].astype('string').str.strip()
ventas['Cod Cliente'] = ventas['Cod Cliente'].astype('string').str.strip()

def _normalize_consolidado_code(series):
    normalized = series.astype('string').str.strip().fillna('')
    return normalized.str.replace(r'\.0+$', '', regex=True)


ventas['Cod Consolidado'] = _normalize_consolidado_code(ventas['Cod Consolidado'])

def _blank_string_mask(series):
    return series.isna() | series.eq('')


def _fill_missing_canal_por_cliente(ventas_df):
    cliente = ventas_df['Cod Cliente']
    canal = ventas_df['Cod Canal Comercial']
    blank_mask = _blank_string_mask(canal)
    known_mask = (~blank_mask) & cliente.notna()

    vacios_originales = int(blank_mask.sum())

    if not known_mask.any():
        print('Cod Canal Comercial vacios originales: 0')
        print('Cod Canal Comercial imputados por cliente: 0')
        print('Cod Canal Comercial vacios restantes: 0')
        print('Clientes con conflicto de canal: 0')
        return ventas_df

    canales_por_cliente = canal.loc[known_mask].groupby(cliente.loc[known_mask], sort=False).nunique(dropna=True)
    clientes_conflictivos = canales_por_cliente[canales_por_cliente > 1].index

    lookup_mask = known_mask & ~cliente.isin(clientes_conflictivos)
    canal_lookup = canal.loc[lookup_mask].groupby(cliente.loc[lookup_mask], sort=False).first()

    fill_values = cliente.loc[blank_mask].map(canal_lookup)
    imputadas = int(fill_values.notna().sum())
    ventas_df.loc[blank_mask, 'Cod Canal Comercial'] = fill_values

    vacios_restantes = int(_blank_string_mask(ventas_df['Cod Canal Comercial']).sum())

    print(f'Cod Canal Comercial vacios originales: {vacios_originales:,}')
    print(f'Cod Canal Comercial imputados por cliente: {imputadas:,}')
    print(f'Cod Canal Comercial vacios restantes: {vacios_restantes:,}')
    print(f'Clientes con conflicto de canal: {len(clientes_conflictivos):,}')

    if len(clientes_conflictivos):
        conflict_mask = known_mask & cliente.isin(clientes_conflictivos)
        ejemplos_conflicto = (
            canal.loc[conflict_mask]
            .groupby(cliente.loc[conflict_mask], sort=False)
            .agg(lambda s: sorted(pd.unique(s).tolist()))
            .head(10)
            .to_dict()
        )
        print('Ejemplos de clientes conflictivos:', ejemplos_conflicto)

    return ventas_df


ventas = _fill_missing_canal_por_cliente(ventas)

clientes_lookup = clientes[['CodCliente', 'Zona', 'Distrito']].copy()
clientes_lookup['CodCliente'] = clientes_lookup['CodCliente'].astype('string').str.strip()
clientes_lookup['Zona'] = clientes_lookup['Zona'].astype('string').str.strip()
clientes_lookup['Distrito'] = clientes_lookup['Distrito'].astype('string').str.strip()
clientes_lookup = clientes_lookup[clientes_lookup['CodCliente'].notna()].copy()

ventas = ventas.merge(
    clientes_lookup.rename(columns={'CodCliente': 'Cod Cliente'}),
    on='Cod Cliente',
    how='left',
    validate='m:1',
)

precios_lista['Material'] = precios_lista['Material'].astype(str).str.strip()
precios_lista['Canal Comercial'] = precios_lista['Canal Comercial'].astype(str).str.strip()
precios_lista['Grandes Cadenas'] = _normalize_consolidado_code(precios_lista['Grandes Cadenas'])
precios_lista['org_ventas_norm'] = (
    precios_lista['Organización ventas']
    .astype('string')
    .str.strip()
    .fillna('')
)
precios_lista = precios_lista[precios_lista['org_ventas_norm'].isin(['', 'PASJ', 'CYSJ'])].copy()
```

# 1. Agregar a cada transacción su precio lista

```python
# --- 2. Versión low memory (SKU + canal + org + cod consolidado) ---
# Evita cruces cartesianos: recorre grupos de precios por organización y
# asigna primero por cod consolidado, usando Grandes Cadenas en blanco como fallback.
def _assign_prices_from_org(out, sales_indices, price_groups, sales_org):
    for key_values, price_rows in price_groups.items():
        if not isinstance(key_values, tuple):
            key_values = (key_values,)

        sales_idx = sales_indices.get(tuple(key_values) + (sales_org,))
        if sales_idx is None or len(sales_idx) == 0:
            continue

        fechas = out.loc[sales_idx, 'Fecha Factura']
        pendientes = out.loc[sales_idx, 'Precio_Lista'].isna()

        for valido_desde, valido_hasta, importe, _ in price_rows.itertuples(index=False, name=None):
            mask = pendientes & fechas.ge(valido_desde) & fechas.le(valido_hasta)
            if mask.any():
                out.loc[mask.index[mask], 'Precio_Lista'] = importe
                pendientes = out.loc[sales_idx, 'Precio_Lista'].isna()
                if not pendientes.any():
                    break


def _build_price_groups(precios_df, org_value):
    prices = precios_df.loc[precios_df['org_ventas_norm'].eq(org_value)].copy()

    exact_prices = prices.loc[prices['Grandes Cadenas'].ne('')].copy()
    fallback_prices = prices.loc[prices['Grandes Cadenas'].eq('')].copy()

    exact_groups = {
        key: (
            sub[['Válido de', 'Fin de validez', 'Importe', '__source_order']]
            .sort_values(['Válido de', '__source_order'], ascending=[False, True])
            .reset_index(drop=True)
        )
        for key, sub in exact_prices.groupby(['Material', 'Canal Comercial', 'Grandes Cadenas'], sort=False, dropna=False)
    }
    fallback_groups = {
        key: (
            sub[['Válido de', 'Fin de validez', 'Importe', '__source_order']]
            .sort_values(['Válido de', '__source_order'], ascending=[False, True])
            .reset_index(drop=True)
        )
        for key, sub in fallback_prices.groupby(['Material', 'Canal Comercial'], sort=False, dropna=False)
    }

    return {
        'exact': exact_groups,
        'fallback': fallback_groups,
    }


def _assign_price_list_low_memory(ventas_df, precios_df):
    out = ventas_df.copy()
    out['Precio_Lista'] = np.nan
    out['org_target'] = out['Distrito'].map({
        'COYHAIQUE': 'CYSJ',
        'PUNTA ARENAS': 'PASJ',
    }).fillna('')

    sales_indices_exact = out.groupby(['Cod SKU', 'Cod Canal Comercial', 'Cod Consolidado', 'org_target'], sort=False, dropna=False).indices
    sales_indices_fallback = out.groupby(['Cod SKU', 'Cod Canal Comercial', 'org_target'], sort=False, dropna=False).indices

    precios_work = precios_df.copy()
    precios_work['__source_order'] = np.arange(len(precios_work))

    price_groups_by_org = {
        org: _build_price_groups(precios_work, org)
        for org in ['CYSJ', 'PASJ', '']
    }

    _assign_prices_from_org(out, sales_indices_exact, price_groups_by_org['CYSJ']['exact'], 'CYSJ')
    _assign_prices_from_org(out, sales_indices_fallback, price_groups_by_org['CYSJ']['fallback'], 'CYSJ')
    _assign_prices_from_org(out, sales_indices_exact, price_groups_by_org['PASJ']['exact'], 'PASJ')
    _assign_prices_from_org(out, sales_indices_fallback, price_groups_by_org['PASJ']['fallback'], 'PASJ')

    for sales_org in ['CYSJ', 'PASJ', '']:
        _assign_prices_from_org(out, sales_indices_exact, price_groups_by_org['']['exact'], sales_org)
        _assign_prices_from_org(out, sales_indices_fallback, price_groups_by_org['']['fallback'], sales_org)

    return out.drop(columns=['org_target'])


merged = _assign_price_list_low_memory(ventas, precios_lista)
```

```python
# Agregar a base de ventas por transacción:
#     - columnas de monto e id para descuentos base, volumen, binario y carta impacto
#     - tabla consolidada de descuentos con todas las reglas posibles
#     - para base/volumen/binario: match exacto por sku+canal+consolidado y fallback por sku+canal


def _as_stripped(series):
    return series.astype("string").str.strip()


def _normalize_consolidado_key(series):
    normalized = series.astype("string").str.strip().fillna("")
    return normalized.str.replace(r"\.0+$", "", regex=True)


def _parse_signed_discount(series):
    cleaned = series.astype("string").str.strip()
    negative_mask = cleaned.str.endswith("-") | cleaned.str.startswith("-")
    cleaned = cleaned.str.replace("-", "", regex=False)
    cleaned = cleaned.str.replace(".", "", regex=False)
    cleaned = cleaned.str.replace(",", ".", regex=False)
    values = pd.to_numeric(cleaned, errors="coerce")
    return values.where(~negative_mask, -values.abs())


def _prepare_discount_frame(df, required_cols, importe_parser, tipo_descuento, rename_map, blank_allowed_cols=None):
    out = df.copy()
    blank_allowed_cols = set(blank_allowed_cols or [])

    for source_col in rename_map:
        if source_col not in out.columns or source_col in ["Válido de ", "A         ", "Importe"]:
            continue

        if source_col == "G.Cadenas":
            out[source_col] = _normalize_consolidado_key(out[source_col])
        else:
            out[source_col] = _as_stripped(out[source_col])

        if source_col in required_cols:
            out = out[out[source_col].notna() & out[source_col].ne("")]
        elif source_col not in blank_allowed_cols:
            out = out[out[source_col].notna() & out[source_col].ne("")]

    out["Válido de "] = pd.to_datetime(out["Válido de "], format="%d.%m.%Y", errors="coerce")
    out["A         "] = pd.to_datetime(out["A         "], format="%d.%m.%Y", errors="coerce")
    out["Importe"] = importe_parser(out["Importe"])
    out["__source_order"] = np.arange(len(out))

    out = out.dropna(subset=required_cols + ["Válido de ", "A         ", "Importe"]).copy()
    out = out.rename(columns=rename_map)

    for col in ["cod_sku", "cod_canal", "cod_consolidado", "cod_cliente"]:
        if col not in out.columns:
            out[col] = pd.NA

    if "cod_sku" in out.columns:
        out["cod_sku"] = _as_stripped(out["cod_sku"])
    if "cod_canal" in out.columns:
        out["cod_canal"] = _as_stripped(out["cod_canal"])
    if "cod_cliente" in out.columns:
        out["cod_cliente"] = _as_stripped(out["cod_cliente"])
    if "cod_consolidado" in out.columns:
        out["cod_consolidado"] = _normalize_consolidado_key(out["cod_consolidado"])

    out["tipo_descuento"] = tipo_descuento

    cols_publicos = [
        "tipo_descuento",
        "cod_sku",
        "cod_canal",
        "cod_consolidado",
        "cod_cliente",
        "fecha_inicio",
        "fecha_fin",
        "monto_descuento",
    ]

    out = (
        out[cols_publicos + ["__source_order"]]
        .drop_duplicates(subset=cols_publicos, keep="first")
        .reset_index(drop=True)
    )

    return out


def _assign_single_discount(ventas_df, descuentos_df, sales_keys, discount_keys, output_col, output_id_col):
    out = ventas_df.copy()
    if output_col not in out.columns:
        out[output_col] = pd.NA
    if output_id_col not in out.columns:
        out[output_id_col] = pd.NA

    lookup = {
        key: sub[["id_descuento", "fecha_inicio", "fecha_fin", "monto_descuento", "__source_order"]].reset_index(drop=True)
        for key, sub in descuentos_df.groupby(discount_keys, sort=False, dropna=False)
    }

    chunks = []
    for key_values, grupo in out.groupby(sales_keys, sort=False, dropna=False):
        if not isinstance(key_values, tuple):
            key_values = (key_values,)

        grupo = grupo.copy()
        grupo["__row_id_tmp"] = grupo.index
        pendientes = grupo[output_col].isna()
        if not pendientes.any():
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        descuentos_sub = lookup.get(tuple(key_values))
        if descuentos_sub is None or descuentos_sub.empty:
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        tmp = grupo.loc[pendientes, ["__row_id_tmp", "Fecha Factura"]].merge(descuentos_sub, how="cross")
        tmp_valid = tmp[
            (tmp["Fecha Factura"] >= tmp["fecha_inicio"]) &
            (tmp["Fecha Factura"] <= tmp["fecha_fin"])
        ]

        if tmp_valid.empty:
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        best = (
            tmp_valid
            .sort_values(["__row_id_tmp", "fecha_inicio", "__source_order", "id_descuento"], ascending=[True, False, True, True])
            .drop_duplicates(subset=["__row_id_tmp"], keep="first")
            [["__row_id_tmp", "monto_descuento", "id_descuento"]]
            .rename(columns={
                "monto_descuento": f"{output_col}__new",
                "id_descuento": f"{output_id_col}__new",
            })
        )

        grupo_out = grupo.merge(best, on="__row_id_tmp", how="left")
        grupo_out[output_col] = grupo_out[output_col].where(grupo_out[output_col].notna(), grupo_out[f"{output_col}__new"])
        grupo_out[output_id_col] = grupo_out[output_id_col].where(grupo_out[output_id_col].notna(), grupo_out[f"{output_id_col}__new"])
        grupo_out = grupo_out.drop(columns=["__row_id_tmp", f"{output_col}__new", f"{output_id_col}__new"])
        chunks.append(grupo_out)

    out = pd.concat(chunks, ignore_index=False).sort_index().reset_index(drop=True)
    out[output_id_col] = out[output_id_col].astype("Int64")
    return out


def _assign_list_discount(ventas_df, descuentos_df, sales_keys, discount_keys, output_col, output_id_col):
    out = ventas_df.copy()
    if output_col not in out.columns:
        out[output_col] = pd.NA
    if output_id_col not in out.columns:
        out[output_id_col] = pd.NA

    lookup = {
        key: sub[["id_descuento", "fecha_inicio", "fecha_fin", "monto_descuento", "__source_order"]].reset_index(drop=True)
        for key, sub in descuentos_df.groupby(discount_keys, sort=False, dropna=False)
    }

    chunks = []
    for key_values, grupo in out.groupby(sales_keys, sort=False, dropna=False):
        if not isinstance(key_values, tuple):
            key_values = (key_values,)

        grupo = grupo.copy()
        grupo["__row_id_tmp"] = grupo.index
        pendientes = grupo[output_col].isna()
        if not pendientes.any():
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        descuentos_sub = lookup.get(tuple(key_values))
        if descuentos_sub is None or descuentos_sub.empty:
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        tmp = grupo.loc[pendientes, ["__row_id_tmp", "Fecha Factura"]].merge(descuentos_sub, how="cross")
        tmp_valid = tmp[
            (tmp["Fecha Factura"] >= tmp["fecha_inicio"]) &
            (tmp["Fecha Factura"] <= tmp["fecha_fin"])
        ]

        if tmp_valid.empty:
            chunks.append(grupo.drop(columns=["__row_id_tmp"]))
            continue

        latest_start = tmp_valid.groupby("__row_id_tmp")["fecha_inicio"].transform("max")
        best = tmp_valid[tmp_valid["fecha_inicio"] == latest_start].copy()
        best = best.sort_values(["__row_id_tmp", "__source_order", "id_descuento"], ascending=[True, True, True])
        best = best.drop_duplicates(subset=["__row_id_tmp", "id_descuento"], keep="first")
        best = (
            best
            .groupby("__row_id_tmp", as_index=False)
            .agg(
                monto_descuento=("monto_descuento", lambda s: [float(x) for x in s.tolist()]),
                id_descuento=("id_descuento", lambda s: [int(x) for x in s.tolist()]),
            )
            .rename(columns={
                "monto_descuento": f"{output_col}__new",
                "id_descuento": f"{output_id_col}__new",
            })
        )

        grupo_out = grupo.merge(best, on="__row_id_tmp", how="left")
        grupo_out[output_col] = grupo_out[output_col].where(grupo_out[output_col].notna(), grupo_out[f"{output_col}__new"])
        grupo_out[output_id_col] = grupo_out[output_id_col].where(grupo_out[output_id_col].notna(), grupo_out[f"{output_id_col}__new"])
        grupo_out = grupo_out.drop(columns=["__row_id_tmp", f"{output_col}__new", f"{output_id_col}__new"])
        chunks.append(grupo_out)

    return pd.concat(chunks, ignore_index=False).sort_index().reset_index(drop=True)


def _assign_single_discount_low_memory(ventas_df, descuentos_df, sales_keys, discount_keys, output_col, output_id_col):
    out = ventas_df.copy()
    if output_col not in out.columns:
        out[output_col] = pd.NA
    if output_id_col not in out.columns:
        out[output_id_col] = pd.NA

    sales_indices = out.groupby(sales_keys, sort=False, dropna=False).indices
    discount_indices = descuentos_df.groupby(discount_keys, sort=False, dropna=False).indices

    for key_values, discount_idx in discount_indices.items():
        if not isinstance(key_values, tuple):
            key_values = (key_values,)

        sales_idx = sales_indices.get(tuple(key_values))
        if sales_idx is None or len(sales_idx) == 0:
            continue

        pendientes = out.loc[sales_idx, output_col].isna()
        if not pendientes.any():
            continue

        descuentos_sub = descuentos_df.iloc[discount_idx][["fecha_inicio", "fecha_fin", "monto_descuento", "id_descuento", "__source_order"]]
        descuentos_sub = descuentos_sub.sort_values(["fecha_inicio", "__source_order", "id_descuento"], ascending=[False, True, True])

        fechas = out.loc[sales_idx, "Fecha Factura"]

        for fecha_inicio, fecha_fin, monto_descuento, id_descuento, _ in descuentos_sub.itertuples(index=False, name=None):
            mask = pendientes & fechas.ge(fecha_inicio) & fechas.le(fecha_fin)
            if mask.any():
                matched_idx = mask.index[mask]
                out.loc[matched_idx, output_col] = monto_descuento
                out.loc[matched_idx, output_id_col] = int(id_descuento)
                pendientes = out.loc[sales_idx, output_col].isna()
                if not pendientes.any():
                    break

    out[output_id_col] = out[output_id_col].astype("Int64")
    return out.sort_index().reset_index(drop=True)


def _split_discount_exact_fallback(descuentos_df):
    exact_mask = descuentos_df["cod_consolidado"].notna() & descuentos_df["cod_consolidado"].ne("")
    return descuentos_df.loc[exact_mask].copy(), descuentos_df.loc[~exact_mask].copy()


merged = merged.drop(
    columns=[
        "dscto_base",
        "id_descuento_base",
        "dscto_volumen",
        "ids_descuento_volumen",
        "dscto_binario",
        "id_descuento_binario",
        "carta_impacto",
        "id_descuento_carta_impacto",
    ],
    errors="ignore",
)
merged["Fecha Factura"] = pd.to_datetime(merged["Fecha Factura"], errors="coerce")
merged["Cod SKU"] = _as_stripped(merged["Cod SKU"])
merged["Cod Canal Comercial"] = _as_stripped(merged["Cod Canal Comercial"])
merged["Cod Consolidado"] = _normalize_consolidado_key(merged["Cod Consolidado"])
merged["Cod Cliente"] = _as_stripped(merged["Cod Cliente"])

dsctos_base = _prepare_discount_frame(
    dsctos_base,
    required_cols=["Material", "C.Comerc."],
    importe_parser=lambda s: pd.to_numeric(s, errors="coerce"),
    tipo_descuento="base",
    rename_map={
        "Material": "cod_sku",
        "C.Comerc.": "cod_canal",
        "G.Cadenas": "cod_consolidado",
        "Válido de ": "fecha_inicio",
        "A         ": "fecha_fin",
        "Importe": "monto_descuento",
    },
    blank_allowed_cols=["G.Cadenas"],
)
dsctos_volumen = _prepare_discount_frame(
    dsctos_volumen,
    required_cols=["Material", "C.Comerc."],
    importe_parser=_parse_signed_discount,
    tipo_descuento="volumen",
    rename_map={
        "Material": "cod_sku",
        "C.Comerc.": "cod_canal",
        "G.Cadenas": "cod_consolidado",
        "Válido de ": "fecha_inicio",
        "A         ": "fecha_fin",
        "Importe": "monto_descuento",
    },
    blank_allowed_cols=["G.Cadenas"],
)
dsctos_binario = _prepare_discount_frame(
    dsctos_binario,
    required_cols=["Material", "C.Comerc."],
    importe_parser=_parse_signed_discount,
    tipo_descuento="binario",
    rename_map={
        "Material": "cod_sku",
        "C.Comerc.": "cod_canal",
        "G.Cadenas": "cod_consolidado",
        "Válido de ": "fecha_inicio",
        "A         ": "fecha_fin",
        "Importe": "monto_descuento",
    },
    blank_allowed_cols=["G.Cadenas"],
)
carta_impacto = _prepare_discount_frame(
    carta_impacto,
    required_cols=["Material", "Cliente"],
    importe_parser=lambda s: pd.to_numeric(s, errors="coerce"),
    tipo_descuento="carta_impacto",
    rename_map={
        "Material": "cod_sku",
        "Cliente": "cod_cliente",
        "Válido de ": "fecha_inicio",
        "A         ": "fecha_fin",
        "Importe": "monto_descuento",
    },
)

consolidado_descuentos = pd.concat(
    [dsctos_base, dsctos_volumen, dsctos_binario, carta_impacto],
    ignore_index=True,
)
consolidado_descuentos = consolidado_descuentos.sort_values(
    [
        "tipo_descuento",
        "cod_sku",
        "cod_canal",
        "cod_consolidado",
        "cod_cliente",
        "fecha_inicio",
        "fecha_fin",
        "monto_descuento",
        "__source_order",
    ],
    kind="stable",
    na_position="last",
).reset_index(drop=True)
consolidado_descuentos["id_descuento"] = np.arange(1, len(consolidado_descuentos) + 1, dtype=np.int64)
consolidado_descuentos = consolidado_descuentos[[
    "id_descuento",
    "tipo_descuento",
    "cod_sku",
    "cod_canal",
    "cod_consolidado",
    "cod_cliente",
    "fecha_inicio",
    "fecha_fin",
    "monto_descuento",
    "__source_order",
]]

output_path_descuentos = "datos/Descuentos históricos/consolidado_descuentos.csv"
consolidado_descuentos.drop(columns=["__source_order"])

dsctos_base = consolidado_descuentos[consolidado_descuentos["tipo_descuento"] == "base"].copy()
dsctos_volumen = consolidado_descuentos[consolidado_descuentos["tipo_descuento"] == "volumen"].copy()
dsctos_binario = consolidado_descuentos[consolidado_descuentos["tipo_descuento"] == "binario"].copy()
carta_impacto = consolidado_descuentos[consolidado_descuentos["tipo_descuento"] == "carta_impacto"].copy()

dsctos_base_exact, dsctos_base_fallback = _split_discount_exact_fallback(dsctos_base)
dsctos_volumen_exact, dsctos_volumen_fallback = _split_discount_exact_fallback(dsctos_volumen)
dsctos_binario_exact, dsctos_binario_fallback = _split_discount_exact_fallback(dsctos_binario)

merged = _assign_single_discount_low_memory(
    merged,
    dsctos_base_exact,
    sales_keys=["Cod SKU", "Cod Canal Comercial", "Cod Consolidado"],
    discount_keys=["cod_sku", "cod_canal", "cod_consolidado"],
    output_col="dscto_base",
    output_id_col="id_descuento_base",
)
merged = _assign_single_discount_low_memory(
    merged,
    dsctos_base_fallback,
    sales_keys=["Cod SKU", "Cod Canal Comercial"],
    discount_keys=["cod_sku", "cod_canal"],
    output_col="dscto_base",
    output_id_col="id_descuento_base",
)
merged = _assign_list_discount(
    merged,
    dsctos_volumen_exact,
    sales_keys=["Cod SKU", "Cod Canal Comercial", "Cod Consolidado"],
    discount_keys=["cod_sku", "cod_canal", "cod_consolidado"],
    output_col="dscto_volumen",
    output_id_col="ids_descuento_volumen",
)
merged = _assign_list_discount(
    merged,
    dsctos_volumen_fallback,
    sales_keys=["Cod SKU", "Cod Canal Comercial"],
    discount_keys=["cod_sku", "cod_canal"],
    output_col="dscto_volumen",
    output_id_col="ids_descuento_volumen",
)
merged = _assign_single_discount_low_memory(
    merged,
    dsctos_binario_exact,
    sales_keys=["Cod SKU", "Cod Canal Comercial", "Cod Consolidado"],
    discount_keys=["cod_sku", "cod_canal", "cod_consolidado"],
    output_col="dscto_binario",
    output_id_col="id_descuento_binario",
)
merged = _assign_single_discount_low_memory(
    merged,
    dsctos_binario_fallback,
    sales_keys=["Cod SKU", "Cod Canal Comercial"],
    discount_keys=["cod_sku", "cod_canal"],
    output_col="dscto_binario",
    output_id_col="id_descuento_binario",
)
merged = _assign_single_discount_low_memory(
    merged,
    carta_impacto,
    sales_keys=["Cod SKU", "Cod Cliente"],
    discount_keys=["cod_sku", "cod_cliente"],
    output_col="carta_impacto",
    output_id_col="id_descuento_carta_impacto",
)

print(f"Tabla consolidada de descuentos guardada en: {output_path_descuentos}")
print(f"Total descuentos consolidados: {len(consolidado_descuentos):,}")
print(f"dscto_base asignado: {merged['dscto_base'].notna().sum():,}")
print(f"dscto_volumen asignado: {merged['dscto_volumen'].notna().sum():,}")
print(f"dscto_binario asignado: {merged['dscto_binario'].notna().sum():,}")
print(f"carta_impacto asignado: {merged['carta_impacto'].notna().sum():,}")

cols_muestra_dsctos = [
    "Fecha Factura",
    "Cod SKU",
    "Cod Canal Comercial",
    "Cod Consolidado",
    "Cod Cliente",
    "dscto_base",
    "id_descuento_base",
    "dscto_volumen",
    "ids_descuento_volumen",
    "dscto_binario",
    "id_descuento_binario",
    "carta_impacto",
    "id_descuento_carta_impacto",
]
merged.loc[
    merged[[
        "id_descuento_base",
        "ids_descuento_volumen",
        "id_descuento_binario",
        "id_descuento_carta_impacto",
    ]].notna().any(axis=1),
    cols_muestra_dsctos,
].head()
```
